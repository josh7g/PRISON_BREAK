## HELICOPTER PRISON BREAKS

This projects seeks to uncover details of prison breaks over the years, specifically from 1971 to 2020.

The main aim of this project is to 

1.Show which year experienced the most prison breaks.

2.Identify the number of successful prison ecapes since 1971

3.Identify which country attempts the most prison breaks.

4.Which country has the most successful prison breaks

```python
#Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
%matplotlib inline
```

```python
#Getting data from wikipedia and loading to a Data frame
data = pd.read_html("https://en.wikipedia.org/wiki/List_of_helicopter_prison_escapes")[1]
data = pd.DataFrame(data)
```

```python
data
#Viewing loaded data
```

## Extracting the years from the Date column

To perform the next operation, we need to extract just the year column to perform the analysis since both date and
month would not be necessary.
I chose to use the datetime module to perform this task

```python
data["Date"] = pd.to_datetime(data["Date"])
data['Date'] = data['Date'].dt.strftime('%Y') #To extract just the year
print(data["Date"])
```

We have sucessfully isolated the years from the Date column. Next, we will see the number of escapes per year.

 ## NUMBER OF ESCAPES PER YEAR

```python
values = data["Date"].value_counts()
print(values)
 ## To show the number of escapes per year
```

Next, we want to show our data on a graph to help others understand clearly what the data is revealing

```python
sns.set(font_scale= 1.0 )
data['Date'].value_counts().plot(kind ='barh', figsize=(10, 8), rot= 0 )
plt.xlabel("Number of attempts")
plt.ylabel("Years")
plt.title("Bar Graph showing the number of attempted Helicopter escapes per year")
plt.show()
```

According to the data, 2007,1986,2001 and 2009 had the most attempted prison escapes with a value of 3 each. 
Let us dive deeper and see which year had the most successful attempts.

To do this, let us first figure out how many successful attempts have been achieved since 1971

## Number of Sucessful attempts

```python
Success = data["Succeeded"].value_counts()
print(Success)
```

This data shows that, there has been 34 successful prison escapes since 1971. This means that, the success rate for prison
escapes according to the data is 70%. 
Let us see a pie chart to give a visual representation of what the Data says

```python
Success = data["Succeeded"].value_counts().plot.pie()
```

## Which year had the most successful attempts?

Now lets us see, which years had the most successful prison escapes

```python
Success = data[data["Succeeded"] == "Yes"]
print(Success[["Date","Succeeded"]])
# This shows which years of a successful prison escape
```

```python
success_frequency = data.groupby([data["Date"],data["Succeeded"] == "Yes"]).count()
success_frequency
# To show the number of successful attempts per year
```

From this we can tell that in the years 1986 and 2009 there were 3 successful prison escapes. This shows that, 2009 and 1986 had the most successsful prison attempts with a value of 3.

Let us move on and see which country has attempted the most prison breaks

## Which country has attempted the most prison breaks?

```python
Country = data["Country"].value_counts()
print(Country)
# To count the number of attempted prison breaks for each country
```

This data shows that France has attempted the most prison breaks with 15, followed by the United States with 8. I am curious to know how many of these attempts were successful 

Let us do one more analysis to find which country has had the most successful prison breaks although I think 
France and the United States are favourites. 

## WHICH COUNTRY HAS THE MOST SUCCESSFUL PRISON ESCAPES?


```python
success_country = data.groupby([data["Country"],data["Succeeded"] == "Yes"]).count()
success_country
```

This is amazing. This data shows that out of the 15 prison escape attempts in France, 11 of them were successful. 
This is a 73% success rate. 
In the United States, out of the 8 prison escape attempts, 6 of them were successful. This is a 75% success rate.
